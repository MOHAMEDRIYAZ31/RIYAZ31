```python
flatten_size_ang=image_size_ang[0]*image_size_ang[1]
flatten_size_ang
```

# Fatten image to vector

```python
for i in range(len_of_image_smi):
    smi_gray[i]=np.ndarray.flatten(smi_gray[i]).reshape(flatten_size_smi,1)
```

```python
for i in range(len_of_image_sad):
    sad_gray[i]=np.ndarray.flatten(sad_gray[i]).reshape(flatten_size_sad,1)
```

```python
for i in range(len_of_image_ang):
    ang_gray[i]=np.ndarray.flatten(ang_gray[i]).reshape(flatten_size_ang,1)
```
